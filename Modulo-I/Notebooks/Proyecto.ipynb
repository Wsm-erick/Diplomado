```python
import pandas as pd
import numpy as np

# Modelado
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

# Métricas
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

# 1) Carga
df = pd.read_csv("data.csv")

# 2) Separar X, y
# Ajusta el nombre exacto si tu CSV difiere (algunas versiones usan 'Target' o 'Status')
target_col = "Target"
X = df.drop(columns=[target_col])
y = df[target_col].astype("category")

# 3) Detectar tipos (puedes hacer una lista manual si prefieres control total)
# Regla: numéricas = int/float; categóricas = object/category/bool
num_cols = X.select_dtypes(include=["int64", "float64", "int32", "float32"]).columns.tolist()
cat_cols = X.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

# (Opcional) si algunas columnas codificadas 0/1 realmente son categóricas ordinales, puedes moverlas manualmente a num_cols/cat_cols

# Transformador de columnas: one-hot a categóricas; escalado SOLO a numéricas
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
    ],
    remainder="drop",
)

# 4) train/test estratificado (mantener proporción de clases)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# 5) Definir tres pipelines (evita fuga de datos al hacer el escalado dentro del pipeline)
pipe_knn = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", KNeighborsClassifier())
])

pipe_log = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", LogisticRegression(max_iter=2000, multi_class="multinomial", n_jobs=None))
])

pipe_tree = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", DecisionTreeClassifier(random_state=42))
])

# 6) Entrenar y evaluar función helper
def evaluar(modelo, nombre):
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1m = f1_score(y_test, y_pred, average="macro")
    print(f"\n=== {nombre} ===")
    print(f"Accuracy: {acc:.4f} | F1-macro: {f1m:.4f}")
    print(classification_report(y_test, y_pred))
    print("Matriz de confusión:")
    print(confusion_matrix(y_test, y_pred, labels=y.cat.categories))

evaluar(pipe_knn, "KNN (k=5 por defecto)")
evaluar(pipe_log, "Regresión Logística (multinomial)")
evaluar(pipe_tree, "Árbol de Decisión")

# 7) GridSearch para KNN (elige k óptimo)
param_grid_knn = {
    "clf__n_neighbors": [3,5,7,9,11,13,15],
    "clf__weights": ["uniform", "distance"],
    "clf__p": [1, 2],  # 1=Manhattan, 2=Euclidiana
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid_knn = GridSearchCV(pipe_knn, param_grid_knn, scoring="f1_macro", cv=cv, n_jobs=-1)
grid_knn.fit(X_train, y_train)

print("\nMejor KNN:", grid_knn.best_params_, "→ F1-macro CV:", grid_knn.best_score_)
# Evaluación final con el mejor KNN en test
y_pred_best = grid_knn.best_estimator_.predict(X_test)
print("\n=== KNN (mejor estimador) ===")
print(f"Accuracy: {accuracy_score(y_test, y_pred_best):.4f} | F1-macro: {f1_score(y_test, y_pred_best, average='macro'):.4f}")
print(classification_report(y_test, y_pred_best))
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred_best, labels=y.cat.categories))
```
